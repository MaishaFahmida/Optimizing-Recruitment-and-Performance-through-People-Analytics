# Data-Driven Recruitment Analysis for Intercity Bus Driver Role
This notebook loads the recruitment data, preprocesses it, runs association rule mining, and explains the top rule with support, confidence, and lift metrics.

```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules


df = pd.read_csv(r'C:\Users\Maisha Fahmida\Desktop\PA\CSV file\fau_transline_recruitment.csv')
print("Initial shape:", df.shape)
df.head()
```

```python

df_intercity = df[df['intercity_bus_driver'] == True].copy()
print("Intercity applications:", df_intercity.shape)
```

```python

for col, thresh in [
    ('experience_years', df_intercity['experience_years'].median()),
    ('previous_companies', df_intercity['previous_companies'].median()),
    ('distance_from_company', df_intercity['distance_from_company'].median())
]:
    df_intercity[col.replace('_years', '_high').replace('previous_companies', 'previous_companies_high').replace('distance_from_company', 'distance_far')] =         (df_intercity[col] >= thresh).astype(int)

df_intercity['safe_driving_high'] = (df_intercity['safe_driving_skills'] >= 75).astype(int)
df_intercity['customer_service_high'] = (df_intercity['customer_service_skills'] >= 75).astype(int)
df_intercity['navigation_high'] = (df_intercity['navigation_skills'] >= 75).astype(int)


df_processed = pd.get_dummies(
    df_intercity,
    columns=['education_level', 'recruitment_strategy'],
    prefix=['edu', 'strategy']
)

df_processed['hired'] = df_processed['hiring_decision'].astype(int)


drop_cols = [
    'experience_years','previous_companies','distance_from_company',
    'safe_driving_skills','customer_service_skills','navigation_skills',
    'hiring_decision','intercity_bus_driver','depot_driver','city_bus_driver'
]
df_processed.drop(columns=drop_cols, inplace=True)
df_processed.dropna(inplace=True)
print("Processed shape:", df_processed.shape)
```

```python

df_bool = df_processed.astype(bool)
print(df_bool.dtypes.value_counts())
```

```python

frequent_itemsets = apriori(df_bool, min_support=0.1, use_colnames=True)

rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.0)
rules_filtered = rules[(rules['lift'] > 1) & (rules['confidence'] > 0.5)]
rules_sorted = rules_filtered.sort_values(['lift','confidence'], ascending=False)
rules_sorted.head(5)
```

```python

first = rules_sorted.iloc[0]
ante = ', '.join(first['antecedents'])
cons = next(iter(first['consequents']))
sup = first['support']
conf = first['confidence']
lf = first['lift']
print(f"Top rule: ({ante}) → ({cons})")
print(f"Support: {sup:.2f} ({sup*100:.0f}%)")
print(f"Confidence: {conf:.2f} ({conf*100:.0f}%)")
print(f"Lift: {lf:.2f}")
```

```python

import matplotlib.pyplot as plt
top5 = rules_sorted.head(5)
labels = top5['antecedents'].astype(str)
fig, ax1 = plt.subplots(figsize=(10,6))

fig.patch.set_facecolor('white')
ax1.set_facecolor('white')
ax1.bar(labels, top5['support'], color='green', label='Support')
ax1.set_xlabel('Antecedents')
ax1.set_ylabel('Support')
ax1.grid(axis='y', linestyle='--', alpha=0.7)
ax2 = ax1.twinx()
ax2.plot(labels, top5['confidence'], marker='o', label='Confidence')
ax2.plot(labels, top5['lift'], marker='x', label='Lift')
ax2.set_ylabel('Confidence & Lift')

ax1.set_xticks(range(len(labels)))
ax1.set_xticklabels(labels, rotation=45, ha='right')
title = (
    'Association Rule Metrics' +
    '\nThe top rule is ' +
    str(rules_sorted.iloc[0]['antecedents']) + ' → ' +
    str(list(rules_sorted.iloc[0]['consequents'])[0])
)
ax1.set_title(title)
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2, loc='upper left')
plt.tight_layout()
plt.show()
```
